```python
import pandas as pd

tbl0 = pd.read_csv("tbl0.tsv", sep="\t")
tbl1 = pd.read_csv("tbl1.tsv", sep="\t")
tbl2 = pd.read_csv("tbl2.tsv", sep="\t")
len(tbl0)
```

```python
len(tbl0.columns)
```

```python
tbl0['_c1'].str[0].value_counts().sort_index()
```

```python
tbl0.groupby(tbl0['_c1'].str[0])['_c2'].mean().sort_index()
```

```python
tbl0.groupby(tbl0["_c1"].str[0])["_c2"].max().sort_index()
```

```python
sorted(tbl1['_c4'].str.upper().unique())
```

```python
tbl0.groupby(tbl0["_c1"].str[0])["_c2"].sum().sort_index()
```

```python
tbl0['suma'] = tbl0['_c0'] + tbl0['_c2']
print(tbl0)
```

```python
tbl0['año'] = tbl0['_c3'].str.split('-').str[1]
tbl0
```

```python
tabla = tbl0.groupby('_c1')['_c2'].agg(list)
tabla = tabla.apply(lambda x: ':'.join(map(str,sorted(x)))).reset_index()
tabla.columns = ["_c1","_c2"]
tabla = tabla.set_index('_c1')
tabla
```

```python
tabla = tbl1.groupby('_c0')['_c4'].agg(lambda x: ','.join(map(str, x))).reset_index()
tabla
```

```python
tabla_p12 = tbl2.copy()
tabla_p12['_c5'] = tabla_p12['_c5a'] + ':' + tabla_p12['_c5b'].astype(str)
tabla_p12 = tabla_p12.groupby('_c0')['_c5'].apply(lambda x: ','.join(map(str,sorted(x)))).reset_index()
tabla_p12
```

```python
tabla_p13 = pd.merge(tbl0,tbl2)
tabla_p13.groupby('_c1')['_c5b'].sum()
```
